```python
#Borrare el Atributo nombre, la intencion es no individualizar al paciente. 
df_health = df_health.drop(labels='Name', axis=1)
```

```python
#Cambiare a 0 los masculinos y 1 los femeninos
df_health['Gender'] = df_health['Gender'].map({'Male': 0, 'Female': 1}).astype(int)
```

```python
#Convierto variable de tipo object en datetime
df_health['Date of Admission'] = pd.to_datetime(df_health['Date of Admission'])
```

```python
#Convierto variable de tipo object en datetime
df_health['Discharge Date'] = pd.to_datetime(df_health['Discharge Date'])
```

```python
df_health['day difference'] = df_health['Discharge Date'] - df_health['Date of Admission']
```
